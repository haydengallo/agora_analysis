```python
### read in abundance data 

### reading in AMANHI Pakistan term abundance data
AMANHI_P_term = pd.read_csv('./cobratoolbox_analysis/Print_relab_AGORA/Metagenomic_abundances_AMANHI_P_term.csv', index_col=0)
AMANHI_P_term.head()


### reading in AMANHI Pakistan pre-term abundance data 

AMANHI_P_pre_term = pd.read_csv('./cobratoolbox_analysis/Print_relab_AGORA/Metagenomic_abundances_AMANHI_P_preterm.csv', index_col=0)
AMANHI_P_pre_term.head()


```

```python
### Sorting by the most prevalent bacteria averaged across samples

sort_average_AMANHI_P_term = AMANHI_P_term.sum(axis=1).sort_values(ascending=False)/AMANHI_P_term.shape[0]
sort_average_AMANHI_P_pre_term = AMANHI_P_pre_term.sum(axis=1).sort_values(ascending=False)/AMANHI_P_pre_term.shape[0]
```

```python
### Take the top 40 on average most prevalent strains per sample for both term and preterm 

AMANHI_P_term_top_40 = AMANHI_P_term.loc[sort_average_AMANHI_P_term.head(40).index,:]
AMANHI_P_pre_term_top_40 = AMANHI_P_pre_term.loc[sort_average_AMANHI_P_pre_term.head(40).index,:]

AMANHI_P_term_top_40.head(20)
AMANHI_P_pre_term_top_40.head(20)
```

```python
filter_species_level = []
filter_species_level_pre_term = []
AMANHI_P_term_new_index =[]
AMANHI_P_pre_term_new_index =[]

for i in range(0, 40):

    if len(AMANHI_P_term_top_40.index[i].split('_')) == 2:
        filter_species_level.append(i)
        AMANHI_P_term_new_index.append('pan'+AMANHI_P_term_top_40.index[i])
        print(AMANHI_P_term_top_40.index[i],'term')

    if len(AMANHI_P_pre_term_top_40.index[i].split('_')) == 2:
        filter_species_level_pre_term.append(i)
        print(AMANHI_P_pre_term_top_40.index[i],'pre')
        AMANHI_P_pre_term_new_index.append('pan'+AMANHI_P_pre_term_top_40.index[i])
        #print(i, AMANHI_P_term_top_40.index[i])
    #if (AMANHI_P_term_top_40.index[i].split('_')[1]) == 'sp':
    #    None
    #else : 
len(filter_species_level)
    #    filter_species_level.append(i)
```

```python
# Filtering for only full species in the top 40 and changing index to work with pan species models


# Term
AMANHI_P_term_top_40_only_species = AMANHI_P_term_top_40.iloc[filter_species_level,:]
AMANHI_P_term_top_40_only_species.index = AMANHI_P_term_new_index

# Pre-term
AMANHI_P_pre_term_top_40_only_species = AMANHI_P_pre_term_top_40.iloc[filter_species_level,:]
AMANHI_P_pre_term_top_40_only_species.index = AMANHI_P_pre_term_new_index
```

```python
list(AMANHI_P_pre_term_top_40_only_species.columns) + list(AMANHI_P_term_top_40_only_species.columns)
```

```python
### Now create group.csv for metadata of pre term and term subjects

sampInfo_AMANHI_P = pd.DataFrame()
sampInfo_AMANHI_P.index = list(AMANHI_P_pre_term_top_40_only_species.columns) + list(AMANHI_P_term_top_40_only_species.columns)
sampInfo_AMANHI_P['Group'] = 'term'
sampInfo_AMANHI_P['Group'][0:len(AMANHI_P_pre_term_top_40_only_species)] = 'preterm'
#sampInfo_AMANHI_P['Group'][len(AMANHI_P_pre_term_top_40_only_species)+1:len(AMANHI_P_term_top_40_only_species)] = 'term'

sampInfo_AMANHI_P.head()

# Saving sample info for AMANHI P

sampInfo_AMANHI_P.to_csv('/Users/haydengallo/Documents/Bucci_Lab/cobratoolbox_analysis/agora_term_preterm_data/sampInfo_AMANHI_P.csv')
```
